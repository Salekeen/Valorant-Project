```python
import pandas as pd
```

```python
cols = ['Team','Mouse','HZ','DPI','Sens','eDPI','Scoped_Sens','Monitor',\
    'MonitorHZ','GPU','Resolution','Mousepad','Keyboard','Headset']
df = pd.read_csv('ProSettings.csv',header=None,index_col=1,names=cols)
```

```python
df.index.name = "Player IGN"
```

#### Fixing Series Types

```python
# replacing the ',' for coversion
df['HZ'] = df['HZ'].str.replace(',','')
df['DPI'] = df['DPI'].str.replace(',','')
df['eDPI'] = df['eDPI'].str.replace(',','')
```

```python
colsToCovert = ['HZ','DPI','Sens','eDPI','Scoped_Sens']
```

```python
# As pandas doesnt support object to numeric conversion
# we first need to convert to str then numeric
for cols in colsToCovert:
    df[cols] = df[cols].astype(str).astype('float32')
```
